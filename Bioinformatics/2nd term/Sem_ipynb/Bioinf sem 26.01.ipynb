## SNP

Ссылка на базу [dbSNP](https://www.ncbi.nlm.nih.gov/snp/), которая находится на NCBI и представляет базу SNP(single nucleotide polimorphism) - однонуклеотидный полиморфизм. Тут еще есть малые вставки, например, и другие различия генома человека. SNP - замена одного нуклеотида в геноме. У нас есть геном человека какой-то сборки , и мы можем сравнить свой геном относительно этого, отличительные места можем назвать полиморфизмами. Какие-то могут играть важную роль в геноме.

Все полиморфизмы обозначаются как rs<число> (rs121912431). Вставим его в строку поиска, нажмем search (ФОТО).

Alleles -показано, какая происходит замена (G меняется на A или C)

В данном случае разные результаты из-за того, что в проектах разная база людей. Если смотреть замены, связанные, например, с цветом кожи, то у них будет различная частота в зависимости от популяции (разных проектов)

Указано, что snp находится в гене, что он меняет аминокислоту (первая строка) (ФОТО)

Не все замены меняют аминокислоту. 

1.   Генетический код избыточен, могут быть замены, но кислота останется та же. 

2.   snp может произойти в некодирующем участке - белок не меняется, но может поменяться регуляторный участок (участок связывания или сигнальную последовательность).

3.   snp может поменять аминокислоту, но при этом с белком ничего не произошло относительно его функции. Когда аминокислота в белке поменялась на похожую, но важное ее свойство для белка не поменялось, то функция белка не поменяется. Если физико-химические свойства аминокислоты не поменялись, то, возможно, его функция не поменяется. Кроме того, обычно белок работает не целиком, а какими-то активными центрами, поэтому если замена произойдет в этой активной зоне, то она окажет большее действие на функциональность белка, чем если замена произошла бы в структурной части белка.

Без клинических данных нельзя определить важность замены. В исследованиях берутся здоровые люди и больные, для них секвенируют геном, находят снипы и пытаются связать данные с болезнью

Для некоторых белков возможна трехмерная структура, по которой можно пытаться определить важность замены.

Ребята посмотрели, в каких местах находятся снипы (ФОТО). Burried - внутри белка, surface - на поверхности. У surface снипы в основном находятся в петлях (вторичные структуры белка - петли, альфа-спирали и бета-листы, соединенные петлями). Тот пример, который мы искали (ФОТО). Мы можем проверять определенный кусок гена до беременности, чтобы избежать большой вероятности тяжелого заболевания. 

Многие белки работают в комплексах, и снипы могут разрушать комплекс (ФОТО)

Могут быть дистабилизирующие полиморфизмы, мешать сборке (белок не свернется правильно), может нарушиться построение белка (у белка есть гидрофобное ядро. Перестройки внутри этого центра могут развернуть весь белок, что повлияет на его функции)

## Пример 2

52 из 374 SNP меняют активность белка. (Смотрим пример 2 из содержания).

Чтобы сбросить PyMOL, идем в Reinitialize - Everything.

Пишем fetch 2PMF. Проверяем мутацию из примера - идем с ncbi omim, ставим SNP и пишем его номер

Есть база SNPedia - в ней много клинических данных. Забьем id сюда тоже, чтобы сравнить данные. SNPedia - свободная база.

Полиморфизм всегда указывается относительно референсного, но в референсном может оказаться минорный аллель, потому что так получилось, что в выборку, из которой составлялся референсный геном, попался минорный аллель. Тот аллель, что в референсном геноме, необязательно является мажорным. Например, референс составлялся в основном на европейцах, и это значит, что для китайцев, например, он может отличаться. Это касается и выявления заболеваний - для китайцев выявление может работать хуже, так как в популяциях разные частоты появления того или иного аллеля.

Ищем 526 позицию - select resi 526. Видим, что данный нуклеотид внутри белка - скорее всего, это активный центр. Нажмем на большой шар в центре - это атом хлора (ФОТО). Раскрасим наш полиморфизм, чтобы увидеть, что он рядом с этим шаром (ФОТО). Сделаем show sticks (ФОТО). Тут можно прочитать, что мы выбрали - D углерод аргинина в позиции 526 от цепи A из объекта 2PMF (ФОТО).

Идем в Wizard - Mutagenesis - Protein, выбираем аминокислоту, меняем на глицин (GLY), жмем Apply - Done. Видим, что у нас освободилось много места (ФОТО). В содержании описана данная замена (см Пример 2) 

Можем найти исходную форму, сравнить с мутантной

Делаем fetch 2ZT5. Справа жмем A(Action) - align - to molecule - 2PMF для выравнивания. Белки хорошо накладываются, так как схожи, кроме одного участка. Это либо мутация, либо просто кристалл получился по-другому. Чтобы определить, мутация ли это, раскрасим C(Color) - spectrum - b-factors. Видим, что часть более красная, чем все остальное (ФОТО). Это значит, что она сильнее болтается. Можем сказать, что это движение из-за температуры и не играет роль. b-factors: чем краснее, тем меньше мы доверяем этому участку структуры. В PDB в Rfree (?) идет темпераутрный компонент.

Если мы сделаем show sticks в том месте, где мы производили мутации, увидим, что оно накладывается, изменения минимальны (ФОТО).

Переименуем A(Action) - rename object - wizrdRtoG

fetch 2PMF (то что изначально с аргинином), затем раскрасим красным место которое мы меняли (ФОТО)

Синее - с глицином, оранжевое - с аргинином (мутированная и вызывающая болезнь) (ФОТО). Они практически схожи, кроме того, что мы обосновываем температурой, изменение в пределах активного центра. Структура белка не изменилась, но в результате снипа поменялась его активность.

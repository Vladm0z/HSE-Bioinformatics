Как находить p-value для таблиц? Тест Фишера и хи-квадрат. Минус теста Фишера - долго считается на больших данных, там в формуле есть факториалы. Поэтому точный тест Фишера на маленьких, в то время как хи-квадрат лучше использовать на больших, на маленьких у него проблемы.

```python
from scipy.stats import chi2_contingency
stat, p, dof, expected =chi2_contingency([[4, 102],[2, 99]],
correction=False)
print(p)
#точный тест Фишера
from scipy import stats
table=[[2,99],[4,102]]
oddsratio, pvalue=stats.fisher_exact(table)
print(pvalue)
print(oddsratio)
```

Можно забить это в калькуляторе, а можно так. У нас p-value получилось 0.4, для точного теста 0.68. То есть эти признаки не являются связанными (p-value большое), то есть локус не является статистически значимым.
